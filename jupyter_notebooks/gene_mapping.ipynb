```r
source("https://bioconductor.org/biocLite.R")
biocLite("org.Hs.eg.db")
```

```r
library(org.Hs.eg.db)
```

```r
# phospho.data <- read.csv("../data/sub_challenge_2_3/breast_cancer_dataset/retrospective_breast_phospho_filtered.txt", 
#                          sep = "\t", colClasses="character")
phospho.data <- read.csv("../data/sub_challenge_2_3/ovarian_cancer_dataset//retrospective_ova_phospho_filtered.txt", 
                         sep = "\t", colClasses="character")
```

```r
gene.ids <- phospho.data[,"Gene_ID"]
```

```r
refSeq <- sapply(gene.ids, function(gene.id) strsplit(gene.id, "[.]")[[1]][1])
sites <- sapply(gene.ids, function(gene.id) strsplit(gene.id, "[:]")[[1]][2])
```

```r
head(refSeq)
```

```r
head(sites)
```

```r
lookup.table <- select(org.Hs.eg.db, keys=keys(org.Hs.eg.db), columns=c("REFSEQ", "SYMBOL", "ALIAS", "GENENAME"))
```

```r
conversion.vector  <- lookup.table$SYMBOL
names(conversion.vector) <- lookup.table$REFSEQ
```

```r
conversion.vector <- conversion.vector[!duplicated(names(conversion.vector))]
```

```r
head(names(conversion.vector), 25)
```

```r
head(conversion.vector[refSeq])
```

```r
phospho.data[,"Gene_ID"] <- paste(conversion.vector[refSeq], sites, sep=":")
```

```r
head(phospho.data$Gene_ID)
```

```r
phospho.data <- phospho.data[!duplicated(phospho.data$Gene_ID),]
```

```r
rownames(phospho.data) <- phospho.data$Gene_ID
```

```r
dim(phospho.data)
```

```r
phospho.data <- subset(phospho.data, select = -c(Gene_ID))
```

```r
write.table(phospho.data, file="../data/sub_challenge_2_3/ovarian_cancer_dataset/phospho_common.txt", sep="\t", )
```
